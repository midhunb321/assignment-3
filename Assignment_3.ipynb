Case Study on Testing of Hypothesis

Question:2,a

```python
import pandas as pd
df2=pd.read_csv('Sales_add.csv')
```

```python
df2
```

```python
df2.columns
```

```python
sales_before = df2['Sales_before_digital_add(in $)']
sales_after = df2['Sales_After_digital_add(in $)']
```

```python
from scipy import stats

t_stat, p_value = stats.ttest_rel(sales_before, sales_after)
```

```python

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant increase in sales after implementing digital marketing.")
else:
    print("Fail to reject the null hypothesis. There is no significant increase in sales after implementing digital marketing.")
```

Qn2:b

```python
df2.columns
```

```python
region = df2['Region']
manager = df2['Manager']
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(region, manager)
chi2, p, dof, expected = chi2_contingency(contingency_table)
```

```python

alpha = 0.05

if p < alpha:
    print("Reject the null hypothesis. Region and Manager are dependent.")
else:
    print("Fail to reject the null hypothesis. Region and Manager are independent.")
```
